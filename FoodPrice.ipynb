```python
df.head()

### Thang đo thuộc tính của:
# Place, ProductName,UmName: norminal
# Month, ProID, UmID :interval scale
# Price, Year: ratio scale
```

```python
#Giá trị TB mỗi loại thực phẩm
rs = df.pivot_table(['Price'],index=['ProductId','ProductName'],aggfunc={'Price':'sum','ProductId':'count'})
rs
```

```python
#Giá trị TB mỗi loại thực phẩm
rs = df.pivot_table(['Price'],index=['ProductId','ProductName'],aggfunc={'Price':'mean'})
rs
```

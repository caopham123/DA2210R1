```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./daily-min-temperatures.csv',encoding='unicode_escape')
df.head()
```

```python
df.describe()
```

```python
x= df.Temp
bins = 45

plt.figure(figsize=(9,5))
plt.hist(x,bins=bins)
plt.title('Temperature Histogram', fontsize = 16)
plt.xlabel('Temperature độ C',fontsize = 14)
plt.show()
```

```python
x= df.Temp
bins = 45

plt.figure(figsize=(9,5))
plt.hist(x,bins=bins, range = (1, 23), histtype = 'step')
plt.title('Temperature Histogram', fontsize = 16)
plt.xlabel('Temperature độ C',fontsize = 14)
plt.show()
```

Vẽ biểu đồ đường thể hiện xu hướng nhiệt độ

```python
df['Date'] = pd.to_datetime(df['Date'])
bounds = ['1/1/1990', '12/31/1990']
bounds = pd.to_datetime(bounds)
df1 = df[(df['Date'] >= bounds[0]) & (df['Date'] <= bounds[1])]

x= df1.Temp
y = df1.Date

plt.figure(figsize=(9,8))
plt.plot(x,y)
plt.title('Temperature Histogram', fontsize = 16)
plt.xlabel('Temperature độ C',fontsize = 14)
plt.ylabel('Date',fontsize = 14)

plt.show()
```

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('./OnlineRetail.csv',encoding='unicode_escape')
df.info()
df.head()

#DL có 541909 dòng và 8 cột
```

```python
# Tạo cột Previous - Quý
df['Day']= pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y\n%H:%M').dt.day
df['Month']= pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y\n%H:%M').dt.month
df['Year']= pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y\n%H:%M').dt.year

df.astype({'Day':'int64', 'Month':'int64','Year':'int64'})

conditions = [(df['Month']>=1) & (df['Month']<=3),
              (df['Month']>=4) & (df['Month']<=6),
              (df['Month']>=7) & (df['Month']<=9),
              (df['Month']>=10) & (df['Month']<=12)
              ]
choices = [1,2,3,4]
df['Previous']= np.select(conditions,choices)
```

```python
#Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice
df['Amount'] = df['Quantity']* df['UnitPrice']
df.head()
```

```python
df1 =df[['Country','Day','Month','Year','Previous']]
df1.head()
```

```python
#Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 
# 5% nếu là ‘France’ ngược lại là 0%.

conditions1 = [(df['Country']=='United Kingdom') & (df['Previous']==4),
               (df['Country']=='France')]
choices1 = ['10%','5%']

df1['Discount'] = np.select(conditions1, choices1, default='0%')
df1.head()
```

```python
df1['Amount']= df['Amount']
df1.astype({'Discount':str})
df1.info()
```
